### Применение Power BI для визуализации и анализа данных с помощью интерактивного дашборда.

##### Схема данных


Данные в таблице Excel хранятся в не нормализованном виде. Присутствует очень много повторений, дублирующихся значений. Данные нужно обработать и загрузить в Power BI. Какую-то обработку можно сделать и в самом Power BI. Большой плюс данного инструмента в том, что можно создать структуру данных и работать уже с многомерным массивом данных. Так называемым OLAP кубом. На данные можно будет смотреть под разными углами, и находить зависимости. Например, мы хотим посмотреть, какая выручка за услугу "Консультация по переводам УКР" была в магазине "20" в январе месяце. Чтобы работать с OLAP кубом, нужно создать структуру данных. Для этого, я выделил одну таблицу фактов "Fact_table", в которой будут содержатся агрегированные данные, которые будут использоваться для составления отчетов, а таблицы измерений будут только описывать хранимые данные. Схема хранилища данных имеет вид "снежинка". Таблицы измерений позволяют избавиться от дублирования данных. Вот, например, в исходных данных в таблице Excel есть лист "Доход". В столбце месяц находится очень много повторяющихся значений. Но такого не должно быть. Все уникальные месяцы должны храниться в отдельной таблице, а на месте текстовых значений должны стоять ключи ссылки на таблицу измерений, где будут храниться только уникальные значения.
